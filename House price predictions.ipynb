```python
columns=['MSZoning','Street','LotShape','LandContour','Utilities','LotConfig','LandSlope',
         'Condition2','BldgType','Condition1','HouseStyle','SaleType',
        'SaleCondition','ExterCond',
         'ExterQual','Foundation','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2',
        'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','Heating','HeatingQC',
         'CentralAir',
         'Electrical','KitchenQual','Functional',
         'FireplaceQu','GarageType','GarageFinish','GarageQual','GarageCond','PavedDrive']
```

```python
df_train.drop('Neighborhood',axis=1,inplace=True)


df_test.drop('Neighborhood',axis=1,inplace=True)

df_train.drop('Id',axis=1,inplace=True)


df_test.drop('Id',axis=1,inplace=True)
```

```python
df_train = df_train.loc[:,~df_train.columns.duplicated()]


df_test = df_test.loc[:,~df_test.columns.duplicated()]
```

```python
train_test=pd.concat([df_train,df_test],axis=0)
```

```python
def fun(columns):
    final_df=train_test
    for col in columns:
        df=pd.get_dummies(final_df[col],drop_first=True)
        final_df.drop(col,axis=1,inplace=True)
        final_df=pd.concat([final_df,df],axis=1)
    return final_df
```

```python
final=fun(columns)
```

```python
final = final.loc[:,~final.columns.duplicated()]     #remove duplicate columns
```

```python
df_train=final.iloc[:1459,:]
df_test=final.iloc[1459:,:]
```

```python
x_train=df_train.drop(['SalePrice'],axis=1)
y_train=df_train['SalePrice']


x_test=df_test.drop(['SalePrice'],axis=1)
```

```python
#train data (Linear regression Algorithm)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
pred=lm.predict(x_test)
```

```python
pred1=pd.DataFrame(pred)
sub_df=pd.read_csv('../input/house-prices-advanced-regression-techniques/sample_submission.csv')
datasets=pd.concat([sub_df['Id'],pred1],axis=1)
datasets.columns=['Id','SalePrice']
datasets.to_csv('sample_submission.csv',index=False)
```
